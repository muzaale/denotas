```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

def scrape_instagram_accounts():
    # URL of the Wikipedia page
    url = "https://en.wikipedia.org/wiki/List_of_most-followed_Instagram_accounts"

    # Send a GET request to the URL and retrieve the content
    response = requests.get(url)
    content = response.text

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(content, "html.parser")

    # Find the table containing the Instagram accounts
    table = soup.find("table", class_="wikitable")

    # Extract the account data from the table
    accounts = []
    rows = table.find_all("tr")[1:]  # Exclude the header row
    for index, row in enumerate(rows, start=1):
        columns = row.find_all("td")
        if len(columns) >= 4:
            rank = columns[0].text.strip()
            account = columns[1].text.strip()
            followers = columns[2].text.strip()
            description = columns[3].text.strip()
            accounts.append([index, rank, account, description])  # Drop the 'Followers' column

    return accounts

# Scrape the Instagram accounts data
accounts_table = scrape_instagram_accounts()

# Print the Instagram accounts table
print(tabulate(accounts_table, headers=["#", "Rank", "Account", "Followers"]))  # Rename the header

```
